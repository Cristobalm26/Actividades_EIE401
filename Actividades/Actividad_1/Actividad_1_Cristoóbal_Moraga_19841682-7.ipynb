### 3.1 Ruido Gaussiano
Debes agregar la señal de ruido a la que creaste previamente. Adicionalmente, debes mostrar una gráfica con el resultado.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal
amplitud = 1.0
frecuencia = 5.0
fase = np.pi / 4
tiempo_final = 1.0
frecuencia_muestreo = 1000
cantidad_puntos = int(tiempo_final * frecuencia_muestreo)

# Vector de tiempo
t = np.linspace(0, tiempo_final, cantidad_puntos)

# Señal senoidal pura sin ruido
senal_pura = amplitud * np.sin(2 * np.pi * frecuencia * t + fase)

# Ruido gaussiano a la señal
media_ruido = 0
desviacion_estandar_ruido = 0.2
ruido = np.random.normal(media_ruido, desviacion_estandar_ruido, cantidad_puntos)
senal_con_ruido = senal_pura + ruido

# Graficar la señal pura y la señal con ruido
plt.figure(figsize=(10, 6))
plt.plot(t, senal_pura, label='Señal Pura', color='blue', linewidth=3)
plt.plot(t, senal_con_ruido, label='Señal con Ruido', color='red', alpha=0.6)
plt.title('Señal Pura vs Señal con Ruido')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```

### 3.2 Ruido Rosa
Investiga en que consiste el ruido rosa, cual es la diferencia del ruido gaussiano y demuéstralo.

El ruido Rosa se define como aquel que tiene una distribución de frecuencias tal que su densidad espectral de potencia es proporcional a la inversa de la frecuencia; lo que origina que su nivel, por bandas de tercio de octava sea constante. Es decir, disminuye en aproximadamente 3 decibelios por octava con el aumento de la frecuencia. Esto significa que las frecuencias más bajas tienen una mayor amplitud que las frecuencias más altas, lo que produce un sonido más equilibrado y natural que el ruido blanco, donde todas las frecuencias tienen la misma amplitud.

Por otra parte, el ruido gaussiano es aquel que presenta una distribución Gauss o normal, donde las variaciones electromagnéticas normalmente son muy pequeñas del orden de los microvoltios siendo despreciable en la mayoria de los sistemas. En otras palabras, la amplitud de las señales en el ruido gaussiano sigue una distribución normal, con la mayoría de los valores concentrados alrededor de la media y una dispersión simétrica de los valores a ambos lados de la media.

Para demostrar la diferencia entre el ruido rosa y el ruido gaussiano

```python
import numpy as np
import matplotlib.pyplot as plt

# Generacion de ruido rosa
def ruido_rosa(n_samples):
    return np.random.randn(n_samples) / np.sqrt(np.arange(1, n_samples + 1))

# Generacion de ruido gaussiano
def ruido_gaussiano(n_samples):
    return np.random.randn(n_samples)

# Número de muestras
n_samples = 1000

# Generar ruido rosa y gaussiano
rosa = ruido_rosa(n_samples)
gaussiano = ruido_gaussiano(n_samples)

# Calculo del espectro de frecuencia utilizando la transformada de Fourier
rosa_fft = np.fft.fft(rosa)
gaussiano_fft = np.fft.fft(gaussiano)

# Densidad espectral de potencia
potencia_rosa = np.abs(rosa_fft)**2
potencia_gaussiano = np.abs(gaussiano_fft)**2

# Escala logarítmica para el eje x
freqs = np.fft.fftfreq(n_samples, d=1)

# Grafica de los espectros de frecuencia
plt.figure(figsize=(12, 6))
plt.plot(freqs[:n_samples//2], 10 * np.log10(potencia_rosa[:n_samples//2]), color='#FF1493', label='Ruido Rosa')
plt.plot(freqs[:n_samples//2], 10 * np.log10(potencia_gaussiano[:n_samples//2]), label='Ruido Gaussiano')
plt.title('Comparacion de los espectros de frecuencia')
plt.xlabel('Frequencia (Hz)')
plt.ylabel('Potencia (dB)')
plt.legend()
plt.grid(True)
plt.show()
```

En el espectro de frecuencia, el ruido rosa exhibe una caída gradual en la potencia a medida que aumenta la frecuencia, mientras que el ruido gaussiano no muestra ninguna tendencia específica y tiene una distribución más uniforme de la potencia en diferentes frecuencias.
